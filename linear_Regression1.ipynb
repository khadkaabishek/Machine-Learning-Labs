P=np.array([200,280,325,290,378,385]) and perform linear regression and find its equations also MSE,R2 score,plot line and predict for 120

```python
import numpy as np
import pandas as pd
```

```python
A=np.array([50,70,75,85,90,101])
P=np.array([200,280,325,390,378,385])
df=pd.DataFrame({"Area":A,"Price":P})
```

```python
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X=df[['Area']]
```

```python
y=df['Price']
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
a=model.intercept_
b=model.coef_
```

```python
print(f"y={a:.2f}+{b[0]:.2f}x")
```

```python
y_predict = model.predict(X)
df['Predicted'] = y_predict
```

```python
df
```

```python
MSE = np.average((df["Price"]-df["Predicted"])**2)
```

```python
MSE
```

```python
#direct method using imports
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse=mean_squared_error(y,df['Predicted'])
```

```python
mse
```

```python
r2_Score = r2_score(y,df['Predicted'])
```

```python
r2_Score
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(X,y)
plt.scatter(X,df['Predicted'
])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Linear Regression")
plt.show()
```

```python
plt.plot(X,df['Predicted'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Linear Regression")
plt.show()
```
