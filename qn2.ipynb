```python
import numpy as np
import pandas as pd
```

```python
data={
    "Years_of_Experience":np.array([4,5,6,3]),
    "Education":np.array(["High School","Bachelor","High School","Masters"]),
    "Salary":np.array([30000,45000,40000,65000])
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.info()
```

```python
#in above , education is of object data type so it need to be preprocessed ie encoding
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l = LabelEncoder()
```

```python
df['Education'
] = l.fit_transform(df['Education'])
```

```python
df
```

```python
df.info()
```

```python
X = df[["Years_of_Experience","Education"]] #also  = df.drop("Salary",axis=1)
y= df["Salary"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
a = model.intercept_
b = model.coef_
```

```python
b
```

```python
print("First Equation")
print(f"y={a:.2f} + {b[0]:.2f}x")
print("Second Equation")
print(f"y={a:.2f} + {b[1]:.2f}x")
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse = mean_squared_error(y,model.predict(X))
r2 = r2_score(y,model.predict(X))
```

```python
mse
```

```python
r2
```

```python
import matplotlib.pyplot as plt
plt.scatter(y,model.predict(X))
plt.plot([y.min(),y.max()],[y.min(),y.max()])
plt.show()
```
