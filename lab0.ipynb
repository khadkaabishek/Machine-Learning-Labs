9. Create a line plot, scatter plot, bar graph and pie chart with labels
using matplotlib for a given series of data


```python
import matplotlib.pyplot as plt

subjects = ["PNS","DSNML","AI","CG","EE"]
marks = [60, 45, 59, 72, 80]

# ---------------- LINE PLOT ----------------
plt.figure(figsize=(8,5))
plt.plot(subjects, marks, marker='o')
plt.title("Line Plot of Subjects vs Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.xticks(rotation=15)
plt.grid(True)
plt.show()


# ---------------- SCATTER PLOT ----------------
plt.figure(figsize=(8,5))
plt.scatter(subjects, marks)
plt.title("Scatter Plot of Subjects vs Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.xticks(rotation=15)
plt.grid(True)
plt.show()


# ---------------- BAR GRAPH ----------------
plt.figure(figsize=(8,5))
plt.bar(subjects, marks)
plt.title("Bar Graph of Subjects vs Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.xticks(rotation=15)
plt.show()


# ---------------- PIE CHART ----------------
plt.figure(figsize=(7,7))
plt.pie(marks, labels=subjects, autopct='%1.1f%%')
plt.title("Pie Chart of Marks Distribution")
plt.show()
```
