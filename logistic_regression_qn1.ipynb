1. Create a dataframe for the following data:
X= 1 2 3 4 5;

    Y =2 5 7 8 9
a) Use alpha/learning rate = 0.01, and find the optimal value of slope and intercept for the
above continuous data
b) Find the optimized equation
c) Determine MSE and R2-score for the above data.

```python
import numpy as np
import pandas as pd
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,5,7,8,9])
```

```python
df = pd.DataFrame({"X":x,"Y":y})
```

```python
X = df['X']
y=df['Y']
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
a=0
b=0
alpha=0.01
n=len(df['X'])
for i in range(0,100):
  y_pred = a+(b*X)
  error = y-y_pred
  mse = mean_squared_error(y,y_pred)

  print(f"MSE = {mse:.3f}")
  # print("R2",r2)
  a_grad = (-2/n)*np.sum(error)
  b_grad = (-2/n)*np.sum(error*X)
  a = a - alpha*a_grad
  b = b - alpha*b_grad
r2 = r2_score(y,y_pred)


```

```python
print(f"Optimal value of a = {a:.3f}")
print(f"Optimal value of b = {b:.3f}")
print(f"MSE = {mse:.3f}")
print(f"R2 = {r2:.3f}")
print(f"Optimized Equation: y ={a:.3f} +{b:.3f}x ")
```
